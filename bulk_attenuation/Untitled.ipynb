```python
from unyt import unyt_array, unyt_quantity

def dTdP(alpha: unyt_quantity, 
         T: unyt_array, 
         rho: unyt_quantity,
         Cp: unyt_quantity) -> unyt_array: 
    return alpha * T / (rho * Cp)
```
